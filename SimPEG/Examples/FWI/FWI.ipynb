```python
n1 = 128; n2 = 128;
h1  = 1.0/n1; h2 = 1.0/n2;
w = 100; m = np.ones(n1*n2)
mbc = 1;
k = w*np.sqrt(w**2 * mbc)
import numpy as np
import SimPEG.utils as utils

mesh = SimPEG.mesh.TensorMesh([h1*np.ones(n1),h2*np.ones(n2)])
```

```python
L1 = (1.0/h1**2) * sp.spdiags((np.ones((n1+1,1))*[1, -2, 1]).T,[-1,0,1],n1+1,n1+1);
L2 = (1.0/h2**2) * sp.spdiags((np.ones((n2+1,1))*[1, -2, 1]).T,[-1,0,1],n2+1,n2+1);

av1 = sp.spdiags((np.ones((n1+1,1))*[0.5, 0.5]).T,[0,1],n1,n1+1);
av2 = sp.spdiags((np.ones((n2+1,1))*[0.5, 0.5]).T,[0,1],n2,n2+1);
```

```python
B1 = utils.spzeros(n1+1,n1+1);
B2 = utils.spzeros(n2+1,n2+1);
B1.dtype = complex
B2.dtype = complex
```

```python
B1[0,0] = 1+2*1j*k*h1; B1[-1,-1] = 1+2*1j*k*h1
B2[0,0] = 1+2*1j*k*h2; B2[-1,-1] = 1+2*1j*k*h2
```

```python
# generate the 2D Laplacian
L = sp.kron(sp.identity(n2+1),L1+B1) + sp.kron(L2+B2,sp.identity(n1+1)) 

# Generate the Mass matrix
Av = sp.kron(av2,av1)
M = utils.sdiag(Av.T*utils.mkvc(m))
A = L + w**2 * M
```

```python
q = np.zeros((n1+1,n2+1)); q[64,64] = 1; q = utils.mkvc(q);
u = sp.linalg.spsolve(A,q);
u = u.reshape((n1+1,n2+1),order='F')

```
